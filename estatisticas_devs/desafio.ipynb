```python
# imports - vejo que no exercicio o plt e importado mas ja vem incluso no pandas e por isso usei diretamente no pandas. Se usasse o plt 
# seriam muitas mais linhas de codigo que penso que nao seja o objetivo (costumizacao)
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicion√°rio de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# criar dataframe
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
# converter datas (string) em formato data sem time
df_faturamento["data_ref"] = pd.to_datetime(df_faturamento['data_ref']).dt.date
```

```python
# verificar se os dados sao datas, depois mostrar dataframe
df_faturamento.info()
df_faturamento
```

```python
# grafico de barras ordenado ascending por data
df_faturamento.sort_values('data_ref', ascending=True).plot.bar('data_ref', 'valor')
```

```python
# grafico lineplot ordenado ascesnding por data
df_faturamento.sort_values('data_ref', ascending=True).plot.line('data_ref', 'valor')
```
